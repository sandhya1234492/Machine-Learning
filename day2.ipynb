```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler

# Step 2: Load Dataset
df = pd.read_csv("Iris.csv")

# Step 3: Display Dataset
print("=" * 50)
print("First Five Rows")
print(df.head())

# Step 4: Dataset Information
print("=" * 50)
print("Dataset Information")
print(df.info())

# Step 5: Missing Values
print("=" * 50)
print("Missing Values")
print(df.isnull().sum())

# Step 6: Fill Missing Numerical Values
if "Age" in df.columns:
    df["Age"] = df["Age"].fillna(df["Age"].mean())

# Step 7: Fill Missing Categorical Values
if "Embarked" in df.columns:
    df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

# Step 8: Drop Cabin Column
if "Cabin" in df.columns:
    df.drop("Cabin", axis=1, inplace=True)

# Step 9: Remove Duplicates
df.drop_duplicates(inplace=True)

# Step 10: Encode Gender
if "Sex" in df.columns:
    encoder = LabelEncoder()
    df["Sex"] = encoder.fit_transform(df["Sex"])

# Step 11: One-Hot Encode Embarked
if "Embarked" in df.columns:
    df = pd.get_dummies(df, columns=["Embarked"], drop_first=True)

# Step 12: Drop Non-Numeric Identifier/Text Columns
columns_to_drop = ["PassengerId", "Name", "Ticket"]
for col in columns_to_drop:
    if col in df.columns:
        df.drop(col, axis=1, inplace=True)

# Step 13: Scale Numerical Columns
numeric_columns = ["Age", "Fare"]
existing_numeric = [col for col in numeric_columns if col in df.columns]

if existing_numeric:
    scaler = StandardScaler()
    df[existing_numeric] = scaler.fit_transform(df[existing_numeric])

# Step 14: Separate Input and Target Variables
X = df.drop("Id", axis=1)
y = df["Id"]

# Step 15: Display Final Dataset Information
print("=" * 50)
print("Processed Dataset")
print(df.head())

print("=" * 50)
print("Input Shape:", X.shape)
print("Target Shape:", y.shape)

print("=" * 50)
print("Data Types")
print(df.dtypes)

print("=" * 50)
print("Final Missing Values")
print(df.isnull().sum())

print("=" * 50)
print("Data Preprocessing Completed Successfully!")
```
