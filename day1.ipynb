```python
import matplotlib.pyplot as plt

# Sample data
ages = [22, 25, 30, 35, 40]
salaries = [30000, 40000, 55000, 60000, 80000]

# Create the 'figure' (canvas) and 'axes' (the plot area)
plt.figure(figsize=(8, 5))  # Width=8, Height=5 inches

# Plot the line/scatter
plt.scatter(ages, salaries, color='blue', label='Salary Data')  # Dots
plt.plot(ages, salaries, color='red', linestyle='--')         # Line
```

```python
# Labels are mandatory!
plt.xlabel("Age (Years)")
plt.ylabel("Salary ($)")
plt.title("Age vs Salary Correlation")
plt.legend()  # Shows the "Salary Data" label.
plt.grid(True)  # Adds a grid for easier reading.
plt.show()  # Renders it.
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split


```

```python
df = pd.read_excel("Workbook1.xlsx")

print("=" * 60)
print("FIRST FIVE RECORDS")
print("=" * 60)
print(df.head())
```

```python
print("\nLAST FIVE RECORDS")
print(df.tail())
```

```python
print("\nDATASET SHAPE")
print(df.shape)

rows, cols = df.shape

print(f"Rows : {rows}")
print(f"Columns : {cols}")
```

```python
print("\nCOLUMN NAMES")
print(df.columns)
```

```python
print("\nDATA TYPES")
print(df.dtypes)
```

```python
print("\nDATASET INFORMATION")
df.info()
```

```python

print("\nSTATISTICAL SUMMARY")
print(df.describe())
```

```python
print("\nMISSING VALUES")
print(df.isnull().sum())
```

```python

print("\nDUPLICATE VALUES")
print(df.duplicated().sum())
```

```python
print("\nUNIQUE values")
print(df["Category"].unique())

print("\nCLASS COUNT")
print(df["Category"].value_counts())

```

```python
print("\nUNIQUE values (State)")
print(df["State"].unique())

print("\nCLASS COUNT (State)")
print(df["State"].value_counts())
```

```python
X = df.drop(columns=["Profit"])

print("\nFEATURES")
print(X.head())
```

```python
y = df["Profit"]

print("\nTARGET")
print(y.head())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20
)
```

```python
print("\nTRAIN TEST SPLIT")

print(f"Training Features : {X_train.shape}")
print(f"Testing Features  : {X_test.shape}")

print(f"Training Labels   : {y_train.shape}")
print(f"Testing Labels    : {y_test.shape}")

```

```python


print("\nTRAINING FEATURES")
print(X_train.head())

print("\nTRAINING LABELS")
print(y_train.head())

```

```python

print("\nTEST FEATURES")
print(X_test.head())

print("\nTEST LABELS")
print(y_test.head())
```

```python
print("\nTRAINING CLASS DISTRIBUTION")
print(y_train.value_counts())

print("\nTESTING CLASS DISTRIBUTION")
print(y_test.value_counts())

```
